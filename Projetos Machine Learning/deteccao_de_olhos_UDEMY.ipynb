```python
import cv2
```

```python
detector_olhos = cv2.CascadeClassifier('/content/drive/MyDrive/PastaCursoVisaoComputacionalUdemy/Cascades/haarcascade_eye.xml')
```

```python
imagem = cv2.imread('/content/drive/MyDrive/PastaCursoVisaoComputacionalUdemy/Images/people1.jpg')
#imagem = cv2.resize(imagem, (800,600))
imagem_cinza =(cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY))
detector_facial = cv2.CascadeClassifier('/content/drive/MyDrive/PastaCursoVisaoComputacionalUdemy/Cascades/haarcascade_frontalface_default.xml')
deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor =1.3, minSize =(30,30))
for x,y,w,h in deteccoes:
  cv2.rectangle(imagem, (x,y), (x+w, y+h), (0,255,0), 2)
deteccoes_olhos = detector_olhos.detectMultiScale(imagem_cinza, scaleFactor = 1.09, minNeighbors=10, minSize=(10,10), maxSize=(50,50))
for (x,y,w,h) in deteccoes_olhos:
  cv2.rectangle(imagem, (x,y), (x+w, y+h), (0,0,255), 2)
from google.colab.patches import cv2_imshow
cv2_imshow(imagem)
```
